### Merge Divvy & Weather Data

```python
weather = pd.read_csv("weather_info.csv", index_col=0)
```

```python
weather.head()
```

```python
# Convert unix -> time
datetime.fromtimestamp(1514726379).strftime("%H:%M:%S")
```

#### Extract month, day, year in Weather and Divvy DF

```python
weather['month'] = pd.DatetimeIndex(weather['date']).month
weather['day'] = pd.DatetimeIndex(weather['date']).day
weather['year'] = pd.DatetimeIndex(weather['date']).year
```

```python
weather.head()
```

```python
divvy.head()
```

```python
splited = divvy['START TIME'].str.split(n=2, expand=True)
```

```python
divvy['START DATE_'] = splited[0]
divvy['START TIME_'] = splited[1]
divvy['START DAYNIGHT_'] = splited[2]
```

```python
divvy['START DATE_'] = pd.to_datetime(divvy['START DATE_'], format='%m/%d/%Y')
divvy['START TIME_'] = pd.to_datetime(divvy['START TIME_'], format='%H:%M:%S')
```

```python
divvy['month'] = divvy['START DATE_'].dt.month
divvy['day'] = divvy['START DATE_'].dt.day
divvy['year'] = divvy['START DATE_'].dt.year
divvy['dayname'] = divvy['START DATE_'].dt.weekday_name
divvy['hour'] = divvy['START TIME_'].dt.hour
```

```python
divvy.loc[divvy['START DAYNIGHT_']=='PM', 'hour'] += 12
```

```python
# Checking datatype of columns that will be used for merge
divvy.dtypes
```

```python
weather.dtypes
```

```python
# merge trip with weather information
merged = pd.merge(left=divvy, right=weather,
                  on=['month', 'day', 'year'], how='inner')
```

```python
# Change all column names to the lowercase.
merged.columns = [x.lower() for x in merged.columns]
```
